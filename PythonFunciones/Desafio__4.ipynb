4 - Crea una funci贸n que reciba dos listas como par谩metros y agrupe los elementos uno a uno de las listas, formando una lista de tuplas de 3 elementos. El primer y segundo elemento de la tupla son los valores en la posici贸n i de las listas y el tercer elemento es la suma de los valores en la posici贸n i de las listas.

* La funci贸n debe incluir un manejo de error indicando el tipo de error generado y devolver como resultado la lista de tuplas. Si las listas enviadas como par谩metro tienen tama帽os diferentes, la funci贸n debe devolver un IndexError con la frase: 'La cantidad de elementos en cada lista es diferente.'.

* Datos para probar la funci贸n:

* Valores sin error:

lista1 = [4, 6, 7, 9, 10]
lista2 = [-4, 6, 8, 7, 9]
Listas con tama帽os diferentes:

lista1 = [4, 6, 7, 9, 10, 4]
lista2 = [-4, 6, 8, 7, 9]
Listas con valores incoherentes:

lista1 = [4, 6, 7, 9, 'A']
lista2 = [-4, 'E', 8, 7, 9]

```python
def agrupar_y_sumar(lista1, lista2):
    resultado = []

    try:
        if len(lista1) != len(lista2):
            raise IndexError("La cantidad de elementos en cada lista es diferente.")

        for i in range(len(lista1)):
            suma = lista1[i] + lista2[i]
            resultado.append((lista1[i], lista2[i], suma))

    except IndexError as e:
        print(f"Error: {e}")
    except TypeError as e:
        print(f"Error de tipo de dato: {type(e).__name__} - {e}")
    except ValueError as e:
        print(f"Error de valor: {type(e).__name__} - {e}")
    finally:
        print("Fin de la ejecuci贸n de la funci贸n.")
        return resultado

# Caso 1: Listas v谩lidas
print("CASO 1 - SIN ERRORES:")
lista1 = [4, 6, 7, 9, 10]
lista2 = [-4, 6, 8, 7, 9]
print(agrupar_y_sumar(lista1, lista2))
print()

# Caso 2: Listas con tama帽os diferentes
print("CASO 2 - TAMAOS DIFERENTES:")
lista1 = [4, 6, 7, 9, 10, 4]
lista2 = [-4, 6, 8, 7, 9]
print(agrupar_y_sumar(lista1, lista2))
print()

# Caso 3: Valores no sumables
print("CASO 3 - VALORES INCOHERENTES:")
lista1 = [4, 6, 7, 9, 'A']
lista2 = [-4, 'E', 8, 7, 9]
print(agrupar_y_sumar(lista1, lista2))
```

Si alguna alternativa en una de las pruebas no est谩 entre las alternativas posibles, debes lanzar un ValueError con el mensaje "La alternativa [alternativa] no es una opci贸n de alternativa v谩lida". El c谩lculo de las 3 notas solo se realizar谩 mediante las entradas con las alternativas A, B, C o D en todas las pruebas. Si no se lanza la excepci贸n, se mostrar谩 una lista con las notas en cada prueba.

* Datos para la prueba del c贸digo:

Respuestas de la prueba:

A continuaci贸n, hay 2 listas de listas que puedes usar como prueba:

* Notas sin excepci贸n:

* Notas con excepci贸n:

```python
def corregir_pruebas(respuestas_correctas, pruebas_estudiantes):
    notas = []
    opciones_validas = {'A', 'B', 'C', 'D'}

    try:
        for prueba in pruebas_estudiantes:
            nota = 0
            for i, respuesta in enumerate(prueba):
                if respuesta not in opciones_validas:
                    raise ValueError(f"La alternativa '{respuesta}' no es una opci贸n de alternativa v谩lida.")
                if respuesta == respuestas_correctas[i]:
                    nota += 1
            notas.append(nota)

    except ValueError as e:
        print(f"Error: {e}")
    finally:
        print("Proceso finalizado.")
        return notas
```

```python
respuestas = ['D', 'A', 'B', 'C', 'A']

# Caso sin excepci贸n
tests_sin_ex = [['D', 'A', 'B', 'C', 'A'],
                ['C', 'A', 'A', 'C', 'A'],
                ['D', 'B', 'A', 'C', 'A']]

# Caso con excepci贸n
tests_con_ex = [['D', 'A', 'B', 'C', 'A'],
                ['C', 'A', 'A', 'E', 'A'],
                ['D', 'B', 'A', 'C', 'A']]

# Prueba 1: sin excepci贸n
print(" Notas sin excepci贸n:")
notas_validas = corregir_pruebas(respuestas, tests_sin_ex)
print("Notas:", notas_validas)
print()

# Prueba 2: con excepci贸n
print(" Notas con excepci贸n:")
notas_invalidas = corregir_pruebas(respuestas, tests_con_ex)
print("Notas:", notas_invalidas)
```

```python
def verificar_puntuacion(lista_palabras):
    puntuaciones = {',', '.', '!', '?'}

    try:
        for palabra in lista_palabras:
            if any(signo in palabra for signo in puntuaciones):
                raise ValueError(f"El texto presenta puntuaciones en la palabra \"{palabra}\"")
        print(f"Todas las palabras est谩n libres de puntuaci贸n.")

    except ValueError as e:
        print(f"Error: {e}")
```
